# Review
This chapter was much smaller and faster than the previous. It covered Cumulative Distribution Functions (CDFs). The concept of a CDF is pretty simple, you just add up the probabilities in a PMF and you result in a CDF. If you have a uniform distribution in your pmf, you can expect a straight line from [0,0] - [1,1] in your cdf. In electrical engineering it was called a step and a ramp function.

Percentiles were another key concept. Percentiles are a normalized rank. These can be spaced evenly into **quartiles**, **quintiles** or even **deciles**! More generally evenly spaced points in a percentile range are called **Quantiles**.

CDFs are particularly good at comparing distributions, especially ones that are close.

If you were a first baby, find your percentile rank in the distribution for first babies. Otherwise use the distribution for others.
```python
import nsfg
import thinkstats2
import thinkplot

preg = nsfg.ReadFemPreg()
live = preg[preg.outcome == 1]
firsts = live[live.birthord == 1]
others = live[live.birthord != 1]

bw_cdf = thinkstats2.Cdf(others.birthwgt_lb)
print(bw_cdf.Prob(9.5)*100)
thinkplot.Cdf(bw_cdf)
thinkplot.Show()
```

## Exercise 2
The numbers generated by random.random are supposed to be uniform between 0 and 1; that is, every value in the range should have the same probability.
Generate 1000 numbers from random.random and plot their PMF and CDF. Is the distribution uniform?

```python
import random
rand_list = [random.random() for i in range(1000)]

pmf = thinkstats2.Pmf(rand_list)
thinkplot.Pmf(pmf)
thinkplot.Show()
```

```python
cdf = thinkstats2.Cdf(pmf)
thinkplot.Cdf(cdf)
thinkplot.Show()
```
